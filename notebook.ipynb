#**Proses Assesing Data**

setelah data dikumpulkan sebanyak 12 tabel

kita akan melakukan tahap pengecekan data apakah terdapat missing,invalid,duplicate,inaccurate,inconsistent,atau ada data yang outlayer.

kita akan menyatukan semua proses assesing ini , dan akan menghasilkan report di akhir

```python
def assesing_data(df,name):
  summary = {
      'nama': name,
      'shape':df.shape,
      'Missing Values': [[column, count] for column, count in df.isna().sum().items() if count > 0] ,
      'Duplicate Data': df.duplicated().sum(),
      'Columns missing': df.columns[df.isna().any()].to_list(),
      'Jumlah Columns missing': df.isna().any().sum(),
      'Data Types': df.dtypes.to_dict()
  }
  return summary
```

```python
datasets = {
    'Aotizhongxin': df_st_atzx,
    'Changping': df_st_cp,
    'Dingling': df_st_dl,
    'Dongsi': df_st_ds,
    'Guanyuan': df_st_gy,
    'Gucheng': df_st_gc,
    'Huairou': df_st_hr,
    'Nongzhanguan': df_st_nzg,
    'Shunyi': df_st_sy,
    'Tiantan': df_st_tt,
    'Wanliu': df_st_wl,
    'Wanshouxigong': df_st_wsxg
}
```

```python
summary = []

for name,df in datasets.items():
  summary.append(assesing_data(df,name))

df = pd.DataFrame(summary)
df
```

#**Cleansing Data**

setelah semua terlihat, kita akan melakukan cleansing data, namun dilihat ternyata missing value di setiap colum cukup banyak dan tidak mungkin untuk di drop value, juga terdapat hari didalam nya sehingga jika di drop akan mendapatkan hilang nya banyak tanggal

sebelum nya saya mencari di internet apa arti dari kolom - kolom singkatan diatas,selain dari tanggal

- PM2.5,	PM10,	SO2,	NO2,	CO,	O3 :jenis jenis partikel udara, data jenis ini cocok untuk dilakukan interpolasi.kenapa? saya rasa cocok untuk data yang tidak akan berubah secara drastis, contoh lain pengukuran berat badan selama 6 bulan seperti ini cocok dilakukan interpolasi jika ada missing value

- TEMP	PRES	DEWP	RAIN,WSPM : jenis jenis ukuran cuaca, data seperti ini cocok dilakukan pengambilan nilai tengah. kenapa tidak interpolasi, karena ukuran cuaca ini tidak datanya berubah drastis namu cenderung stabil, oleh karena itu mean bisa menjadi solusi terbaik
- Rain,WD : curah hujan dan arah angin, saya akan implementasi Imputation Foward/Backward Fill dengan cara ambil data sebelumnya untuk curah hujan dan angin ambil data selanjutnya

Metode Interpolasi linear

```python
column_particle = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
df_st_atzx[column_particle] = df_st_atzx[column_particle].interpolate(method='linear') #Aotizhongxin
df_st_cp[column_particle] = df_st_cp[column_particle].interpolate(method='linear') #Changping
df_st_dl[column_particle] = df_st_dl[column_particle].interpolate(method='linear') #Dingling
df_st_ds[column_particle] = df_st_ds[column_particle].interpolate(method='linear') #Dongsi
df_st_gy[column_particle] = df_st_gy[column_particle].interpolate(method='linear') #Guanyuan
df_st_gc[column_particle] = df_st_gc[column_particle].interpolate(method='linear') #Gucheng
df_st_hr[column_particle] = df_st_hr[column_particle].interpolate(method='linear') #Huairou
df_st_nzg[column_particle] = df_st_nzg[column_particle].interpolate(method='linear') #Nongzhanguan
df_st_sy[column_particle] = df_st_sy[column_particle].interpolate(method='linear') #Shunyi
df_st_tt[column_particle] = df_st_tt[column_particle].interpolate(method='linear') #Tiantan
df_st_wl[column_particle] = df_st_wl[column_particle].interpolate(method='linear') #Wanliu
df_st_wsxg[column_particle] = df_st_wsxg[column_particle].interpolate(method='linear') #Wanshouxigong

print("Replacement missing value to linear method value Success")
```

Metode Interpolasi Mean

```python
column_weather = ['TEMP', 'PRES', 'DEWP', 'RAIN','WSPM']

df_st_atzx[column_weather] = df_st_atzx[column_weather].fillna(df_st_atzx[column_weather].mean()) # Aotizhongxin
df_st_cp[column_weather] = df_st_cp[column_weather].fillna(df_st_cp[column_weather].mean())       # Changping
df_st_dl[column_weather] = df_st_dl[column_weather].fillna(df_st_dl[column_weather].mean())       # Dingling
df_st_ds[column_weather] = df_st_ds[column_weather].fillna(df_st_ds[column_weather].mean())       # Dongsi
df_st_gy[column_weather] = df_st_gy[column_weather].fillna(df_st_gy[column_weather].mean())       # Guanyuan
df_st_gc[column_weather] = df_st_gc[column_weather].fillna(df_st_gc[column_weather].mean())       # Gucheng
df_st_hr[column_weather] = df_st_hr[column_weather].fillna(df_st_hr[column_weather].mean())       # Huairou
df_st_nzg[column_weather] = df_st_nzg[column_weather].fillna(df_st_nzg[column_weather].mean())    # Nongzhanguan
df_st_sy[column_weather] = df_st_sy[column_weather].fillna(df_st_sy[column_weather].mean())       # Shunyi
df_st_tt[column_weather] = df_st_tt[column_weather].fillna(df_st_tt[column_weather].mean())       # Tiantan
df_st_wl[column_weather] = df_st_wl[column_weather].fillna(df_st_wl[column_weather].mean())       # Wanliu
df_st_wsxg[column_weather] = df_st_wsxg[column_weather].fillna(df_st_wsxg[column_weather].mean()) # Wanshouxigong


print("Replacement missing value to mean value Success")
```

Metode Backward/Forward Fill
namun untuk hujan saya akan asumsikan jika nilai kosong saya akan isi 0 artinya tidak ada hujan lokasi itu

```python
# fill 0 untuk hujan
df_st_atzx.RAIN.fillna(value = 0.0, inplace=True)   # Aotizhongxin
df_st_cp.RAIN.fillna(value = 0.0, inplace=True)       # Changping
df_st_dl.RAIN.fillna(value = 0.0, inplace=True)       # Dingling
df_st_ds.RAIN.fillna(value = 0.0, inplace=True)       # Dongsi
df_st_gy.RAIN.fillna(value = 0.0, inplace=True)       # Guanyuan
df_st_gc.RAIN.fillna(value = 0.0, inplace=True)       # Gucheng
df_st_hr.RAIN.fillna(value = 0.0, inplace=True)       # Huairou
df_st_nzg.RAIN.fillna(value = 0.0, inplace=True)     # Nongzhanguan
df_st_sy.RAIN.fillna(value = 0.0, inplace=True)       # Shunyi
df_st_tt.RAIN.fillna(value = 0.0, inplace=True)       # Tiantan
df_st_wl.RAIN.fillna(value = 0.0, inplace=True)       # Wanliu
df_st_wsxg.RAIN.fillna(value = 0.0, inplace=True)   # Wanshouxigong

print("Success fillna with fill 0")

# Forward
df_st_atzx['wd'] = df_st_atzx['wd'].ffill()   # Aotizhongxin
df_st_cp['wd'] = df_st_cp['wd'].ffill()       # Changping
df_st_dl['wd'] = df_st_dl['wd'].ffill()       # Dingling
df_st_ds['wd'] = df_st_ds['wd'].ffill()       # Dongsi
df_st_gy['wd'] = df_st_gy['wd'].ffill()       # Guanyuan
df_st_gc['wd'] = df_st_gc['wd'].ffill()       # Gucheng
df_st_hr['wd'] = df_st_hr['wd'].ffill()       # Huairou
df_st_nzg['wd'] = df_st_nzg['wd'].ffill()     # Nongzhanguan
df_st_sy['wd'] = df_st_sy['wd'].ffill()       # Shunyi
df_st_tt['wd'] = df_st_tt['wd'].ffill()       # Tiantan
df_st_wl['wd'] = df_st_wl['wd'].ffill()       # Wanliu
df_st_wsxg['wd'] = df_st_wsxg['wd'].ffill()   # Wanshouxigong

print("Success fillna with Foward method")
```

```python
corr_column = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3','TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

correlation_ds = df_st_ds[corr_column].corr()
correlation_ds
```
